```

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/mnt/data/Noida_AQIBulletins.csv')

# Convert date column
df['date'] = pd.to_datetime(df['date'])

# Show first 5 rows
df.head()

# Plot 1: Time-Series Trend of Index Value
plt.figure(figsize=(10,5))
plt.plot(df['date'], df['Index Value'])
plt.xlabel("Date")
plt.ylabel("Index Value")
plt.title("Time-Series Trend of Index Value (Noida)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Missing & Abnormal Value Analysis
print("Missing Values:", df['Index Value'].isna().sum())
abnormal = df[(df['Index Value'] < 0) | (df['Index Value'] > 600)]
abnormal.head()

# Plot 2: Distribution of Prominent Pollutant
plt.figure(figsize=(10,5))
df['Prominent Pollutant'].value_counts().plot(kind='bar')
plt.xlabel("Prominent Pollutant")
plt.ylabel("Count")
plt.title("Distribution of Prominent Pollutant (Noida)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
